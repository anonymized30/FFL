```python

#Experiment3: Poisoning attacks in EigenTrust Enivironment(First task) - Malicious selection/detection

import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = [i+1 for i in range(30)]
selected_malicious = np.asarray([4, 4, 4, 4, 2, 2, 1, 1, 2, 2, 5, 5, 2, 2, 1, 1, 1, 1, 0, 0, 1, 1, 3, 3, 0, 0, 1, 1, 2, 2])
detected_malicious = np.asarray([3, 3, 4, 3, 2, 1, 1, 1, 1, 1, 0, 0, 2, 1, 0, 1, 0, 1, 0, 0, 1, 1, 2, 3, 0, 0, 1, 1, 1, 2])

x = np.arange(len(labels))  # the label locations
width = 0.5  # the width of the bars

fig, ax = fig, ax = plt.subplots(figsize=(10, 5), dpi= 300, facecolor='w', edgecolor='k')
rects1 = ax.bar(x, selected_malicious, width, label='Selected malicious peers')
rects2 = ax.bar(x, detected_malicious, width, label='Detected malicious peers')
ax.plot(x, selected_malicious/max(selected_malicious), linewidth=1.5, label='Selected ratio', color='gray', marker='*', linestyle='--')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Count')
ax.set_title('Malicious peers selection/detection in the first training task')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.xlabel("Global training epoch")
plt.ylim(0, 5)

fig.tight_layout()

plt.show()
```
